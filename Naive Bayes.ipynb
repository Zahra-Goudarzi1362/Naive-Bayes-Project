```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
 
```

```python
data = pd.read_csv('E:\Momeni-Machne Learning class momeni 21 01 1402\Machine Learning\dataset/diabetes.csv')
```

```python
data
```

```python
# As In our  data set Pregnancies is discrete we should delete this
# Becasue Gaussian Naive Bayes can use for continuos dfeatures. 
df = pd.DataFrame(data.drop('Pregnancies',axis=1))
df
```

```python
# In plots it seems to be missing value
df1 = df[df['SkinThickness']>0]
df1 = df1[df1['SkinThickness']<90]
df1
```

```python
df2=df1.dropna(how='any')
df2.reset_index(drop=True,inplace=True)
```

```python
df2
```

```python
x = pd.DataFrame(df2, columns=['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']).values
y = df2.Outcome.values.reshape(-1,1)
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.35, random_state=0)
```

```python
clf = GaussianNB()
clf.fit(x_train,y_train.ravel())
y_pred = clf.predict(x_test)
```

```python
print('Accuracy:', metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
clf.classes_
```

```python
confusion_matrix(y, clf.predict(x))
```

```python
cm = confusion_matrix(y, clf.predict(x))

fig, ax = plt.subplots(figsize=(8,8))
ax.imshow(cm)
ax.grid(False)
ax.xaxis.set(ticks=(0,1),ticklabels=('Predicted 0s', 'Predicted 1s'))
ax.yaxis.set(ticks=(0,1),ticklabels=('Actual 0s', 'Actual 1s'))
ax.set_ylim(1.5, -0.5)
for i in range (2):
    for j in range(2):
        ax.text(j,i,cm[i,j], ha='center', va='center', color = 'red')
plt.show()
```

```python
print(classification_report(y, clf.predict(x)))
```
